```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import kmapper as km
import kmapper.evaluate as evaluate
from sklearn import cluster
```

```python
data = np.loadtxt('two_insert_cir.csv', delimiter=',')
projected_data = data[:, 0]
mapper = km.KeplerMapper(verbose=0)
```

```python
def objective_function(X):
    n = int(round(float(X[0])))
    a = float(X[1])
    try:
        cover = km.Cover(n_cubes=n, perc_overlap=a)
        graph = mapper.map(projected_data, data,
                           clusterer=cluster.DBSCAN(eps=0.5, min_samples=3), cover=cover)
        tsc = evaluate.compute_SC_adj(data, projected_data, graph, cover,
                                      type='k', cluster=cluster.DBSCAN(eps=0.5, min_samples=3))
        return -tsc
    except Exception:
        return 3.0

from mealpy import FloatVar
problem = {
    "obj_func": objective_function,
    "bounds": [FloatVar(lb=5, ub=25, name="n"), FloatVar(lb=0.1, ub=0.7, name="a")],
    "minmax": "min",
}
results = run_optimization(problem, 'results_output_two_i_cir.txt')
```

```python
n, a = 25, 0.12
cover = km.Cover(n_cubes=n, perc_overlap=a)
graph = mapper.map(projected_data, data, cluster.DBSCAN(eps=0.5, min_samples=3), cover=cover)
draw_graph(graph, projected_data, '2ic.png')
evaluate.compute_SC_adj(data, projected_data, graph, cover,
                        type='k', cluster=cluster.DBSCAN(eps=0.5, min_samples=3))
```
